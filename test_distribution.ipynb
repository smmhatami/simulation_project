```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Exponential distribution

```python
# draw probability density function of exponential distribution
x = np.linspace(0, 200, 10000)
y = (1/30)*np.exp(-x/30)
plt.plot(x, y)
plt.show()
```

```python
def get_exponential_varites(lamda, size):
    R_i = np.random.uniform(0, 1, size=size)
    data = [(-1/lamda)*np.log(1 - r) for r in R_i]
    return data
```

```python
# draw histogram of exponential distribution
exp_variates_reverse_method = get_exponential_varites(1/30, 10000)
exp_variates_numpy = np.random.exponential(scale=30, size=10000)
plt.figure(figsize=(10, 10))
plt.subplot(211)
plt.hist(exp_variates_reverse_method, bins=150, density=True)
plt.plot(x, y)
plt.title("Reverse Method")
plt.subplot(212)
plt.hist(exp_variates_numpy, bins=150, density=True)
plt.plot(x, y)
plt.title("Numpy")
plt.show()
```

# Normal distribution 

```python
# draw probability density function of normal distribution
x = np.linspace(0, 80, 10000)
mean = 40
variance = 36
y = (1/np.sqrt(2*np.pi*variance))*np.exp(-(x-mean)**2/(2*variance))
plt.plot(x, y)
plt.show()

```

```python
normal_variates_numpy = np.random.normal(loc=40, scale=6, size=10000)
# plt.figure(figsize=(10, 10))
plt.hist(normal_variates_numpy, bins=100, density=True)
plt.plot(x, y)
plt.title("Numpy")
plt.show()
```

# Gamma distribution

```python
# draw probability density function of gamma distribution
x = np.linspace(0, 5, 10000)
beta = 2
theta = 3
y = (beta*theta)**beta*x**(beta-1)*np.exp(-beta*theta*x)/math.gamma(beta)
plt.plot(x, y)
plt.show()
```

```python
gamma_variates_numpy = np.random.gamma(shape=beta, scale=1/(2*theta), size=10000)

plt.hist(gamma_variates_numpy, bins=100, density=True)
plt.plot(x, y)
plt.title("Numpy")
plt.show()
```
